## Skala Radial

```python
# Import Library Tkinter dan math
from tkinter import *
import math
import pandas as pd

## Load data strike
strike = pd.read_csv("strike1.txt")                                             ## Memanggil data strike1.txt ke variabel strike
print(strike)
strike_list = list(strike['Strike'])                                            ## Mengubah dataframe strike dengan kolom 'Strike' menjadi variabel strike_list (kalau di c++ itu array)
print('List Strike:', strike_list)

## Menentukan interval
interval = []                                                                    ## List kosong interval
for k in range(0, 25):                                                           ## Loop untuk membuat 24 interval, jika interval setiap 10 derajat maka bisa diganti 36 interval
    interval.append(k * 15)                                                      ## Mengisi list kosong interval
    print(f'Invertal ke-{k}: {interval[k]}')
print("List Interval:", interval)

## Menentukan Frekuensi dan Persentase Frekuensi dengan list kosong
frekuensi = []                                                                    ## List kosong frekuensi
persentasefrekuensi = []                                                          ## List kosong presentasefrekuensi

## For loop untuk membuat frekuensi ke-i selalu bernilai 0 sebagai nilai awal
for k in range(0, 24):
    frekuensi.append(k * 0)                                                       ## Mengisi list kosong frekuensi dengan nilai 0
print("List Frekuensi awal:", frekuensi)

## For loop untuk membuat Presentase frekuensi ke-i sesuai dengan jumlah frekuensinya di setiap interval
for i in range(0, 24):                                                            ## For Loop untuk mengulang i sesuai dengan banyaknya interval
    for j in range(0, 100):                                                       ## For Loop (Loop dalam Loop) untuk mengulang j sesuai dengan banyaknya strike
        if strike_list[j] >= interval[i] and strike_list[j] < interval[i+1]:      ## If statement untuk menghitung jika nilai strike lebih besar dari interval sebelumnya dan strike lebih kecil dari interval setelahnya
            frekuensi[i] = frekuensi[i] + 1                                       ## Maka nilai frekuensi ke-i diganti dengan pertambahan jumlah frekuensinya sesuai dengan interval
        interval[24] = 0
    persentasefrekuensi.append((frekuensi[i] / 100) * 100)                        ## Menghitung persentase frekuensi yang kemudian di append (dimasukan) ke list kosong presentase frekuensi
    print(f'Persentase Frekuensi ke-{i}: {persentasefrekuensi[i]}')
print("List Persentase Frekuensi:", persentasefrekuensi)

## Menentukan Maxdata untuk membuat jarijari
maxdata = persentasefrekuensi[0]              ## Membuat max data merupakan presentasefrekuensi list pertama (presetasefrekuensi[0]) sebagai awalan          
for i in range(0, 24):                        ## Membuat loopingan yang membandingkan persentase frekuensi dengan maxdata awal
    if maxdata <= persentasefrekuensi[i]:     ## Membuat If statement jika nilai maxdata <= persentase frekuensi ke - i
        maxdata = persentasefrekuensi[i]      ## Maka nilai max datanya sama dengan nilai persentase frekuensi ke - i tersebut, jika tidak maka tetap pakai maxdata sebelumnya
print(f'Max Data yang dipakai: {maxdata}')

# Membuat variabel
## Posisi koordinat, panjang sisi, dan skala
posxa = 350                                      ## Inisiasi variabel posxa dengan nilai 400
posya = 350                                      ## Inisiasi variabel posya dengan nilai 400
jari2 = maxdata                                  ## Inisiasi variabel jari2 dengan nilai maxdata
lebar = 250                                      ## Inisiasi variabel lebar dengan nilai 250
skala = lebar / maxdata                          ## Inisiasi variabel skala dengan nilai lebar/maxdata
phi = 3.141592                                   ## Inisiasi variabel phi dengan nilai 3.141592
awalx = 100                                      ## Inisiasi variabel awalx untuk lingkaran utama = 100
awaly = 100                                      ## Inisiasi variabel awaly untuk lingkaran utama = 100
akhirx = 600                                     ## Inisiasi variabel akhirx untuk lingkaran utama = 600
akhiry = 600                                     ## Inisiasi variabel akhiry untuk lingkaran utama = 600

# Pembuatan Skala Radial (Grafik Roset)
## Main Window
root = Tk()
root.title('Skala Radial (Grafik Roset)')
root.geometry('700x700')

## Canvasa
canvas = Canvas(root, height = 700, width = 700, bg = 'white')

## Membuat garis horisontal dan vertikal di titik tengah
line = canvas.create_line(posxa - lebar, posya, posxa + lebar, posya, fill="blue")   ## Membuat garis horizontal                                        ## Membuat garis B ke C
line = canvas.create_line(posxa, posya - lebar, posxa, posya + lebar, fill="blue")   ## Membuat garis verikal
circle = canvas.create_oval(awalx, awaly, akhirx, akhiry)                            ## Membuat base lingkaran roset

## Menggambar gari-garis Interval dan teks Interval yang sudah ditentukan
for i in range(0, 24):
    x1 = posxa + ((jari2 - 0.5) * math.sin((i*15) * phi / 180) * skala)              ## Posisi x1 untuk line garis-garis interval
    y1 = posya - ((jari2 - 0.5) * math.cos((i*15) * phi / 180) * skala)              ## Posisi y1 untuk line garis-garis interval
    
    x2 = posxa + ((jari2 + 0.5) * math.sin((i*15) * phi / 180) * skala)              ## Posisi x2 untuk line garis-garis interval
    y2 = posya - ((jari2 + 0.5) * math.cos((i*15) * phi / 180) * skala)              ## Posisi y2 untuk line garis-garis interval
    
    x3 = posxa + ((jari2 + 1) * math.sin((i * 15) * phi / 180) * skala)              ## Posisi x3 untuk meletakan teks interval
    y3 = posya - ((jari2 + 1) * math.cos((i * 15) * phi / 180) * skala)              ## Posisi y3 untuk meletakan teks interval
    
    angka = i * 15
    
    line = canvas.create_line(x1, y1, x2, y2, fill="blue")                           ## Membuat garis-garis interval sepanjang lingkaran roset
    text = canvas.create_text(x3,y3,text=angka)                                      ## Membuat teks angka interval sepanjang lingkaran roset

## Plot data
for i in range(0, 24):
    jarijari = persentasefrekuensi[i] / maxdata * jari2 * skala                                                                                                                      ## Membuat jarijari untuk setiap arc yang sesuai presentase strike setiap interval
    arc = canvas.create_arc(awalx + 250 - jarijari, awaly + 250 - jarijari, akhirx - 250 + jarijari, akhiry - 250 + jarijari, start=90 - interval[i+1], extent=15, fill="red")       ## Membuat arc sesuai persentase strike setiap interval
    ## Note: penggunaan + - 250 itu membuat lingkaran arc menjadi titik nol titik pusat, kemudian ditambahkan kebutuhan jari-jari (konsepnya sama seperti radiusnya c++)

    
# Membuat kolom nama dan nim
text = canvas.create_text(10,10,text="Skala Radial (Grafik Roset)",anchor=W)     ## menamakan judul
text = canvas.create_text(10,25,text="Ponijo 190912837",anchor=W)                ## nama dan NIM
    
canvas.pack()
root.mainloop()
```

# Codingan C++ (Jangan di run)

```python
#include <iostream>     // Doni Maulana Syahputra 118150100 Shift B Geokom
#include <math.h>       // Skala radial minggu 6
#include <graphics.h>
#include <fstream>
#include <string>

using namespace std;
int main ()
{
    initwindow (800, 800, "Skala Radial");

    //menginput file yang udh dibuat ke dalam program
    ifstream file1;
    file1.open("strike1.txt");

    //konversi karakter ke angka
    string data;
    int i = 0;
    float strike [100]; //deklarasi strike berupa float sbg array (kumpulan data) kapasitas 100

    while (getline(file1, data))  //input nilai pada file1 sbg data
    {
      stringstream bb(data);    //mengubah karakter mjd float
      bb >> strike [i];         //input bb strike yg punya kapasitas 100
      i++;
    }
    file1.close();                 //menutup file

    //menampilkan data file1
    for (i=0; i<=99; i++)
    {
        cout << strike [i] << endl;
    }

    //menentukan interval
    float interval [24];            //deklarasi interval berupa array
    for (int k=0; k<=23; k++)         //nilai k merupakan pengali untuk interval
    {
        interval [k] = k * 15;          //interval kelipatan 15
        printf ("\ninterval : %f", interval [k]);
    }


    //menentukan frekuensi
    float frekuensi [24];           //deklarasi frekuensi berupa array
    float persentasefrekuensi [24]; //deklarasi persentase frekuensi
    for (i=0;i<=23;i++)
    {
        frekuensi [i]=0;
    }
    for (i = 0; i<=23;i++)
    {
        for (int j = 0; j <= 99; j++)     //deklarasi jumlah strike
        {
            if (strike[j] >= interval [i] && strike [j] < interval [i+1])
            {
                frekuensi [i]++;
            }
            interval [24] = 0;
        }
        persentasefrekuensi [i] = (frekuensi [i] / 100) * 100;
        printf ("\npersen frekuensi:%f", persentasefrekuensi[i]);
    }
    // menentukan jarijari
    float maxdata = persentasefrekuensi [0];
    for (i=0; i<=23; i++)
    {
        if (maxdata <= persentasefrekuensi [i])
            maxdata = persentasefrekuensi [i];
        printf ("\nmaxdata: %f", maxdata);
    }

    //membuat grafik roset
    float jari2 = maxdata;
    int posxa = 400;
    int posya = 400;
    float lebar = 250;

    moveto (posxa, posya);
    line (posxa - lebar, posya, posxa + lebar, posya);
    line (posxa, posya - lebar, posxa, posya + lebar);
    float skala = lebar / maxdata;
    circle (posxa, posya, maxdata * skala);


    // membuat interval
    for (i=0; i<=23; i++)
    {
        double phi=3.141592;

        float x1 = posxa + ((jari2 - 0.5) * sin((i*15) * phi / 180) * skala);
        float y1 = posya - ((jari2 - 0.5) * cos((i*15) * phi / 180) * skala);

        float x2 = posxa + ((jari2 + 0.5) * sin((i*15) * phi / 180) * skala);
        float y2 = posya - ((jari2 + 0.5) * cos((i*15) * phi / 180) * skala);

        float x3 = posxa + ((jari2 + 2) * sin((i * 15) * phi / 180) * skala);
        float y3 = posya - ((jari2 + 3) * cos((i * 15) * phi / 180) * skala);

        char a [5];
        float angka=i * 15;
        snprintf(a, 4, "%f", angka);
        outtextxy (x3, y3, a);

        line (x1, y1, x2, y2);
    }
    //mengeplot data
    for (i=0; i<=23; i++)
    {
        float jarijari = persentasefrekuensi[i] / maxdata * jari2 * skala;
        setfillstyle (5, 1);
        pieslice (posxa, posya, 90 - interval [i+1], 90 - interval[i], jarijari);
    }

    getch ();
    closegraph ();
    return 0;
}
```
